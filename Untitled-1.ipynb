```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1. Read the original image in color
img = cv2.imread("images/tf2_engineer.jpg")

# 2. Get image dimensions and calculate the center coordinates (y_c, x_c)
height, width, _ = img.shape
center_y = height // 2
center_x = width // 2

# 3. Get the BGR pixel value at the center of the original image
original_center_pixel = img[center_y, center_x]

# 4. Define the rectangle patch properties
patch_height = 30
patch_width = 40
top_left = (center_x - patch_width // 2, center_y - patch_height // 2)
bottom_right = (center_x + patch_width // 2, center_y + patch_height // 2)
patch_color = (168, 158, 50)  # Hex #329ea8 converted to BGR format

# 5. Draw the rectangle patch on a copy of the original image
img_with_patch = img.copy()
cv2.rectangle(img_with_patch, top_left, bottom_right, patch_color, thickness=-1)  # filled rectangleee

# 6. Get the new BGR pixel value at the center (which now lies inside the patch)
new_center_pixel = img_with_patch[center_y, center_x]

# 7. Display the image with the color patch (convert BGR to RGB for correct colors in matplotlib)
img_rgb = cv2.cvtColor(img_with_patch, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Image with Color Patch")
plt.axis("off")
plt.show()

# 8. Print out relevant information
print(f"Image size: {width}x{height}")
print(f"Center coordinates: ({center_y}, {center_x})")
print(f"Original center BGR value: {original_center_pixel}")
print(f"New center BGR value (after patch): {new_center_pixel}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the image in grayscale mode
img = cv2.imread("images/pout.tif", cv2.IMREAD_GRAYSCALE)

# 2. Apply Log Transformation
# Convert image to float64 and add 1 to avoid log(0)
c = 255 / np.log(1 + np.max(img))
log_transformed = c * np.log(1 + img.astype(np.float64))
log_transformed = np.array(log_transformed, dtype=np.uint8)

# 3. Apply Inverse Log Transformation
c_inv = 255 / (np.exp(np.max(img)/255) - 1)
inv_log_transformed = c_inv * (np.exp(img.astype(np.float64) / 255) - 1)
inv_log_transformed = np.clip(inv_log_transformed, 0, 255)  # ensure values are in [0,255]
inv_log_transformed = np.array(inv_log_transformed, dtype=np.uint8)

# 4. Apply Inverse Log Transformation on Log Transformed Image
inv_log_on_log = c_inv * (np.exp(log_transformed.astype(np.float64) / 255) - 1)
inv_log_on_log = np.clip(inv_log_on_log, 0, 255)
inv_log_on_log = np.array(inv_log_on_log, dtype=np.uint8)

# 5. Display the images side-by-side
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(log_transformed, cmap='gray')
plt.title("Log Transformed")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(inv_log_transformed, cmap='gray')
plt.title("Inverse Log Transformed")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(inv_log_on_log, cmap='gray')
plt.title("Inverse Log on Log Image")
plt.axis('off')

plt.tight_layout()
plt.show()
#STEP2 The log transformation enhances the details in the darker regions of the image by expanding the lower intensity values. This makes dark areas more visible and improves the contrast in shadows, while compressing the higher intensity values.
#STEP4 Applying the inverse log transform on the log-transformed image attempts to recover the original image. The result closely resembles the original, but slight differences may appear due to rounding and clipping during transformations. This confirms that the log and inverse log transforms are approximately inverse operations.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the image in grayscale
img = cv2.imread("images/moon.tif", cv2.IMREAD_GRAYSCALE)

# 2. Define spatial domain unsharp masking function
def unsharp_mask_spatial(image, k, kernel_size=5):
    # Apply Gaussian blur to the image
    blurred = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
    # Calculate the mask by subtracting the blurred image from the original
    mask = image - blurred
    # Add the scaled mask back to the original image to sharpen
    sharpened = image + k * mask
    # Clip values to valid range and convert to uint8
    sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
    return sharpened

# 3. Define ideal high-pass filter in frequency domain
def ideal_high_pass_filter(shape, D0):
    P, Q = shape
    H = np.zeros((P, Q))
    center_x, center_y = P//2, Q//2
    for u in range(P):
        for v in range(Q):
            # Calculate distance from center of frequency rectangle
            D = np.sqrt((u - center_x)**2 + (v - center_y)**2)
            # Set filter value: 0 inside radius D0, 1 outside (high-pass)
            H[u,v] = 0 if D <= D0 else 1
    return H

# 4. Define frequency domain unsharp masking function
def unsharp_mask_frequency(image, k, D0=30):
    # Compute the 2D Fourier Transform of the image
    F = np.fft.fft2(image)
    F_shift = np.fft.fftshift(F)

    # Create ideal high-pass filter
    H = ideal_high_pass_filter(image.shape, D0)
    # Apply the filter with scaling factor k
    G_shift = (1 + k * H) * F_shift

    # Compute the inverse Fourier Transform to get the sharpened image
    G = np.fft.ifftshift(G_shift)
    g = np.fft.ifft2(G)
    g = np.real(g)
    # Clip values to valid range and convert to uint8
    g = np.clip(g, 0, 255).astype(np.uint8)
    return g

# 5. Test different k values for sharpening
k_values = [0.2, 0.5, 1.0]

# 6. Plot original and sharpened images side by side
plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

for i, k in enumerate(k_values):
    # Apply spatial domain unsharp masking
    spatial_result = unsharp_mask_spatial(img, k)
    # Apply frequency domain unsharp masking
    freq_result = unsharp_mask_frequency(img, k)

    # Show spatial domain results
    plt.subplot(3, 3, 3*i+2)
    plt.imshow(spatial_result, cmap='gray')
    plt.title(f"Spatial k={k}")
    plt.axis('off')

    # Show frequency domain results
    plt.subplot(3, 3, 3*i+3)
    plt.imshow(freq_result, cmap='gray')
    plt.title(f"Frequency k={k}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the image in grayscale
img = cv2.imread("images/pcb.tif", cv2.IMREAD_GRAYSCALE)

# 2. Display the original image and its histogram side by side
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original PCB Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(img.ravel(), bins=256)
plt.title("Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.show()

# 3. Noise analysis explanation
print("Noise Analysis:")
print("- Salt-and-pepper noise typically creates peaks at the two extremes of the histogram.")
print("- Gaussian noise appears as a more spread out, bell-shaped distribution in the histogram.")

# 4. Noise removal using filters
median_filtered = cv2.medianBlur(img, 3)               # Median filter to reduce salt-and-pepper noise
gaussian_filtered = cv2.GaussianBlur(median_filtered, (3,3), 0)  # Gaussian filter to smooth the image

# 5. Display the results after filtering
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter Applied")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title("Median + Gaussian Filter Applied")
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the image
img = cv2.imread("images/pollen.tif", cv2.IMREAD_GRAYSCALE)

# 2. Problem interpretation and evidence (statistical and visual)
print("Original Image Statistics:")
print(f"- Minimum pixel value: {np.min(img)}")
print(f"- Maximum pixel value: {np.max(img)}")
print(f"- Mean pixel value: {np.mean(img):.2f}")
print(f"- Standard deviation: {np.std(img):.2f}")

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(img.ravel(), bins=256)
plt.title("Original Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Problem Interpretation:
print("\nInterpretation:")
print("- The image is noisy, likely contains salt-and-pepper noise.")
print("- There are peaks at extreme values in the histogram.")
print("- High standard deviation indicates strong contrast or noise intensity.")

# 3. Two different approaches (median and gaussian)

# Approach 1: Median filter (good for salt-and-pepper noise)
filtered_median = cv2.medianBlur(img, 3)

# Approach 2: Gaussian filter (general smoothing and low-frequency noise)
filtered_gaussian = cv2.GaussianBlur(img, (5, 5), 0)

# 4. Display the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(filtered_median, cmap='gray')
plt.title("Median Filtered")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(filtered_gaussian, cmap='gray')
plt.title("Gaussian Filtered")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.hist(img.ravel(), bins=256)
plt.title("Original Histogram")

plt.subplot(2, 3, 5)
plt.hist(filtered_median.ravel(), bins=256)
plt.title("Median Histogram")

plt.subplot(2, 3, 6)
plt.hist(filtered_gaussian.ravel(), bins=256)
plt.title("Gaussian Histogram")

plt.tight_layout()
plt.show()

# 5. Evidence and Comparison
def show_stats(image, name):
    print(f"\n{name} Statistics:")
    print(f"- Minimum: {np.min(image)}")
    print(f"- Maximum: {np.max(image)}")
    print(f"- Mean: {np.mean(image):.2f}")
    print(f"- Standard Deviation: {np.std(image):.2f}")

show_stats(filtered_median, "Median Filtered Image")
show_stats(filtered_gaussian, "Gaussian Filtered Image")
```
